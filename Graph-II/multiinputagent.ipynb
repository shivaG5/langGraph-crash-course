```python
from typing import TypedDict,List
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    values:List[int]
    name:str
    result:str
     #3 diff inputs
```

```python
def process_value(state: AgentState)->AgentState:
    """This function handles multiple different input"""
    print(state)
    
    state["result"]=f"Hi there {state['name']}! Your sum is ={sum(state['values'])}"
    print(state)
    return state

```

```python
graph=StateGraph(AgentState)
graph.add_node("process",process_value)
graph.set_entry_point("process") 
graph.set_finish_point("process")
app=graph.compile()
```

```python
from typing import TypedDict,List
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    numbers:List
    name:str
    ao1:str
    res:str
    
def Arp1(state:AgentState)->AgentState:
    """You are an agent that will do the arthimetic operations"""
    if state['ao1']=="+":
        answer1=sum(state['numbers'])
        
    elif state["ao1"]=="*":
        answer1=1
        for numbe in state['numbers']:
            answer1*=numbe
    
    state['res']=f"Hey {state['name']},this is the answer you asked {answer1}" 
    return state
    
```

```python
mol=StateGraph(AgentState)
mol.add_node("operations",Arp1)
mol.set_entry_point("operations")
mol.set_finish_point("operations")
ts=mol.compile()
```

```python
final_answer=ts.invoke({
    "numbers":[1,1,5],
    "name":"John",
    "ao1":"+"
})
```

```python
from typing import TypedDict
from langgraph.graph import StateGraph

class AgentState(TypedDict):
    val3:List
    name:str
    ao3:str
    res:str
    
def arth_agent(state:AgentState)->AgentState:
    """You an simple arthimetic operation"""
    if state['ao3']=="+":
        ans3=sum(state['val3'])
    elif state['ao3']=="*":
        ans3=1
        for num in state['val3']:
            ans3*=num
            
    state['res']=f"Hey hello mr{state['name']} your answer is {ans3}"
            
    return state
```

```python
class AgentState(TypedDict):
    nums:list
    name:str
    ao4:str
    res:str
    
def subdiv(state:AgentState)->AgentState:
    """you are subtraction and dividing agent"""
    if state["ao4"]=="/":
        ans=state["nums"][0]
        
        for num in state['nums']:
            ans/=num
            
    elif state["ao4"]=="-":
        ans=state["nums"][0]
        for num in state["nums"][1:]:
            ans-=num
            
    state["res"]=f"hey mr {state['name']} , you answer is {ans}"
    return state

```

```python
ho=StateGraph(AgentState)
ho.add_node("sub",subdiv)
ho.set_entry_point("sub")
ho.set_finish_point("sub")
fm=ho.compile()
```

```python
final_answer2=fm.invoke({
    'nums':[100,2,5],
    'name':"Luffy",
    'ao4':'-'
    
})
```
